```python
# Numpy
import numpy as np
from numpy import array

# Matplot
import matplotlib.pyplot as plt
%matplotlib inline

# Others
import os 
import pandas as pd
import operator
import glob
```

```python
def count(res):
    res = res[7:]
    list_to_build = []
    i = 7
    while(pd.isnull([res['Ontario Traffic Inc - Bicycle Counts'][i]])):
        i+=1
    j = 0 ## volume

    while(i<=len(res['Unnamed: 2'])+6):
        if(not(pd.isnull(res['Ontario Traffic Inc - Bicycle Counts'][i]))):
            time = res['Ontario Traffic Inc - Bicycle Counts'][i]
            vol = 0
            list_to_build.append((time,vol))
            j = 0 #update vol to be 0
        else:
            j+=1 #add one vol
            #print(list_to_build)
            time = list_to_build[i-1-7][0]
            vol = j+1
            list_to_build.append((time,vol))
            #print(list_to_build)
        i+=1
        
    return list_to_build

def add_time_vol(res):
    raw_list = count(res)
    build_list = []
    for i in range(len(raw_list)):
        time_temp = raw_list[i][0]
        vol = raw_list[i][1]
        if(i!=0):
            if(time_temp != raw_list[i-1][0]):
                build_list.append([raw_list[i-1][0],raw_list[i-1][1]])
    build_list.append([raw_list[len(raw_list)-1][0],raw_list[len(raw_list)-1][1]])
    return build_list    
```

```python
def read_first_intersection(filename):
#filename = '54514_Sussex Ave WB from Spadina Ave to Huron St.xls'
    if 'from ' in filename:
        index_from = filename.index('from ')+5
        index_end = filename.index('to ')
        short = filename[6:index_from-6]+'_'+filename[index_from:index_end-1]
    return short
def read_second_intersection(filename):
#filename = '54514_Sussex Ave WB from Spadina Ave to Huron St.xls'
    if 'from ' in filename:
        index_from = filename.index('from ')+5
        index_end = filename.index('to ')
        short = filename[6:index_from-6]+'_'+filename[index_end+3:-4]
    return short

def full_list_for_every_file(location,filename):
    os.chdir(location)
    bike = pd.read_excel(filename)
    time_vol_list = add_time_vol(bike)
    date = bike['Unnamed: 6'][1][:5]
    intersection1 = read_first_intersection(filename)
    intersection2 = read_second_intersection(filename)
    #print(date) 09/23
    for i in range(len(time_vol_list)):
        time_vol_list[i].append(date)
        time_vol_list[i].append(intersection1)
        time_vol_list[i].append(intersection2)
    return time_vol_list
```

```python
def read_all_files(location):
    os.chdir(location)
    fold_list = []
    #print(fold_location)
    for filename in os.listdir(os.getcwd()):
        #print(filename)
        file_list = full_list_for_every_file(location,filename)
        fold_list.append(file_list)
    return fold_list
```
